# Quantiles
*Touching statistics*

> Percentiles: 
    - q50 means that 50% of your dataset is below that number. That is the median value.
    - q90 means that 90% of your dataset is below that number.
    - q10 means that 10% of your dataset is below that number.

> Calculating the q50 

```python
# Your code here!
df_animals.BodyWt.median()
```

```python
df_animals.loc[df_animals.BodyWt>df_animals.BodyWt.median(), :]
```

```python
df_animals.BodyWt.quantile(0.1)
```

```python
df_animals.BodyWt.quantile([0.1, 0.2, 0.3])
```

## Binarizing (categorizing) data

```python
# Your code here!
df_animals['BodyWt_Interval'] = pd.qcut(df_animals.BodyWt, 4, ['P', 'M', 'G', 'GG'])
```

```python
df_animals.BodyWt_Interval.value_counts(normalize=True)
```

```python
# Your code here!

df_animals['BodyWt_Interval2'] = pd.cut(df_animals.BodyWt, 4)
```

```python
# Your code here!
q10 = df_animals.BodyWt.quantile(0.1)
```

```python
df_animals.query(f'BodyWt < {q10}')
```

## Numpy where

```python
import numpy as np
```

```python
# Your code here!
q95 = df_animals.BodyWt.quantile(0.95)
np.where(df_animals.BodyWt>q95, 'SUPER OVERWEIGHT', 'NORMAL')
```

> np.where(condition is true, then this value, else this value)

```python
# Your code here!
q50 = df_animals.BodyWt.quantile(0.50)
q85 = df_animals.BodyWt.quantile(0.85)
q95 = df_animals.BodyWt.quantile(0.95)
df_animals['class'] = np.where(df_animals['BodyWt'] > q95, 'SUPER-OVERWEIGHT',
                           np.where(df_animals['BodyWt'] > q85, 'ALMOST-SUPER-OVERWEIGHT', 
                                   np.where(df_animals['BodyWt'] > q50,'QUASI-ALMOST-SUPER-OVERWEIGHT','NORMAL')))
```
