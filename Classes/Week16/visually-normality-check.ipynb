```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
from scipy import stats
import numpy as np
```

## Probability Plot
* Plotting distribution quantiles against each other.
> https://towardsdatascience.com/explaining-probability-plots-9e5c5d304703

### Theory

```python
normal_dist = np.random.normal(size=1000)
```

```python
sns.distplot(normal_dist)
```

```python
fig, ax = plt.subplots(1,1, figsize=(6,6))
stats.probplot(normal_dist, plot=ax);
```

```python
fig, axs = plt.subplots(1,3, figsize=(20,5))

stats.probplot(normal_dist, plot=axs[1])

sns.distplot(normal_dist, color='red', bins=20, kde=False, ax=axs[0])
sns.distplot(normal_dist, color='blue', bins=20, kde=False, ax=axs[2])
```

```python
mono_dist = np.random.random(size=1000)
norm_dist_1 = np.random.normal(loc=np.mean(mono_dist), scale=np.std(mono_dist), size=1000)

beta_dist = np.random.beta(a=10, b=1, size=1000)
norm_dist_2 = np.random.normal(loc=np.mean(beta_dist), scale=np.std(beta_dist), size=1000)
```

```python
fig, axs = plt.subplots(1,3, figsize=(20,5))

stats.probplot(mono_dist, plot=axs[1])
sns.distplot(norm_dist_1, color='red', bins=20, kde=False, ax=axs[0])
sns.distplot(mono_dist, color='blue', bins=20, kde=False, ax=axs[2])
```

```python
fig, axs = plt.subplots(1,3, figsize=(20,5))

stats.probplot(beta_dist, plot=axs[1])

sns.distplot(norm_dist_2, color='red', bins=20, kde=False, ax=axs[0])
sns.distplot(beta_dist, color='blue', bins=20, kde=False, ax=axs[2])
```

### Real Data

```python
df = pd.read_csv('data/winequality-red.csv')
```

```python
df.describe().T[['50%', 'mean', 'std']]
```

```python
sns.distplot(df['fixed acidity'])
```

```python
df.skew().sort_values()
```

### free sulfur dioxide

```python
loc = np.mean(df['free sulfur dioxide'])
scale = np.std(df['free sulfur dioxide'])
size = len(df['free sulfur dioxide'])

dist_normal = np.random.normal(loc=loc, scale=scale, size=size)
```

```python
fig, axs = plt.subplots(1,3, figsize=(20,5))

stats.probplot(df['free sulfur dioxide'], plot=axs[1])

sns.distplot(dist_normal, color='red', bins=20, kde=False, ax=axs[0])
sns.distplot(df['free sulfur dioxide'], color='blue', bins=20, kde=False, ax=axs[2])
```

### density

```python
loc = np.mean(df['density'])
scale = np.std(df['density'])
size = len(df['density'])

dist_normal = np.random.normal(loc=loc, scale=scale, size=size)
```

```python
fig, axs = plt.subplots(1,3, figsize=(20,5))

stats.probplot(df['density'], plot=axs[1])

sns.distplot(dist_normal, color='red', bins=20, kde=False, ax=axs[0])
sns.distplot(df['density'], color='blue', bins=20, kde=False, ax=axs[2])
```
