## Dropping columns
> `axis=1` or

> `columns = ['name_of_column_to_drop1', 'name_of_column_to_drop2' ,...]`

### Drop columns based on condition

```python
# Your code here!
mask = (df.isnull().sum()/len(df))>= 0.5
columns_to_drop = mask[mask == True].index
columns_to_drop
```

```python
df.drop(columns_to_drop, axis=1, inplace = True)
```

```python
fig, ax = plt.subplots(figsize=(10,5))
sns.heatmap(~df.isnull().head(1000), cmap = 'coolwarm')
```

### Fill with some value

```python
# Your code here!
df['coluna_nan'] = np.nan
df
```

```python
df.coluna_nan.fillna(-1, inplace = True)
df
```

```python
cylinders_mean = df.cylinders.mean()
df.cylinders.fillna(cylinders_mean)
```

```python
group_fuel_type = df.groupby('fuelType')
for key_ in group_fuel_type.groups.keys():
    temp = group_fuel_type.get_group(key_)
    temp.cylinders.fillna(temp.cylinders.mean(), inplace=True)
```

## Checking duplicates rows
    
>    - Dropping fully duplicate row
>    - Subset
>    - `.duplicated()`

```python
frutas = pd.DataFrame({
    'fruta':['laranja', 'laranja', 'mamÃ£o', 'laranja'],
    'cidade':['Rio de Janeiro', 'Atibaia', 'Campinas', 'Rio de Janeiro']
})
frutas
```

```python
# Your code here!
frutas.duplicated() #verifica se tem registro (linha) duplicado em todas a colunas
```

```python
frutas.duplicated(subset=['fruta']) #verifica se tem linha duplicada apenas na coluna 'fruta'
```

```python
frutas.drop_duplicates()
```

```python
frutas.drop_duplicates(keep=False)
```

```python
frutas.drop_duplicates(subset=['fruta'])
```

```python
frutas.drop_duplicates(subset=['fruta'], keep=False)
```
