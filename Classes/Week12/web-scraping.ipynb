## Find the `table` tag whose `class=sortable wikitable`

```python
# Your code here!
results = soup.find_all('table', attrs={'class': 'sortable wikitable'})
results[0].text
```

## get the table header `<th>`

```python
# Your code here!
th_results = list()

for result in results[0].find_all('th'):
    data = result.text.strip()
    th_results.append(data)
    
th_results
```

## get the table data `<td>`

```python
# Your code here!
td_results = list()

for result in results[0].find_all('td'):
    data = result.text.strip()
    td_results.append(data)
    
td_results
```

```python
np.array(td_results).reshape((int(len(td_results)/4), 4))
```

```python
pd.DataFrame(np.array(td_results).reshape((int(len(td_results)/4), 4)), columns=th_results)
```

# Creating a spider

## Find the book name, price and stock availability as a pandas dataframe.

http://books.toscrape.com/

```python
# Your code here!
page = 10
url = f'http://books.toscrape.com/catalogue/page-{page}.html'
url
```

```python
page = 1

url = f'http://books.toscrape.com/catalogue/page-{page}.html'
response = requests.get(url)
html = response.content
```

```python
soup = BeautifulSoup(html)
```

```python
for h3 in soup.find_all('h3'):
    for a in h3.find_all('a'):
        print(a['title'])
```

```python
titles_list = [a['title'] for h3 in soup.find_all('h3') for a in h3.find_all('a')]
titles_list
```

```python
prices_list = [float(p.text.replace('£', '')) for p in soup.find_all('p', attrs={'class' : 'price_color'})]
prices_list
```

```python
stock_list = [p.text.strip() for p in soup.find_all('p', attrs={'class' : 'instock availability'})]
stock_list
```

```python
dfs = list()

for page in range (1, 51):
    
    url = f'http://books.toscrape.com/catalogue/page-{page}.html'
    
    response = requests.get(url)
    html = response.content
    soup = BeautifulSoup(html)
    
    titles_list = [a['title'] for h3 in soup.find_all('h3') for a in h3.find_all('a')]
    prices_list = [float(p.text.replace('£', '')) for p in soup.find_all('p', attrs={'class' : 'price_color'})]
    stock_list = [p.text.strip() for p in soup.find_all('p', attrs={'class' : 'instock availability'})]

    dfs.append(pd.DataFrame({'title':titles_list, 'price':prices_list, 'stock':stock_list, 'page':page}))
    
```

```python
df = pd.concat(dfs, ignore_index=True)
df
```
